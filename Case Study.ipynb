```python
#Import the pandas library

import numpy as np
import pandas as pd
```

```python
#import company files

companies = pd.read_csv('https://cdn.upgrad.com/UpGrad/temp/d934844e-5182-4b58-b896-4ba2a499aa57/companies.txt', sep = '\t', encoding = 'latin')
companies
```

```python
#import rounds2 file
rounds2 = pd.read_csv('https://cdn.upgrad.com/UpGrad/temp/4c3b5ed0-e5dc-4838-89a2-173d8707d857/rounds2.csv', sep = "," , encoding = 'latin')
rounds2
```

```python
#decode the values of the field name
companies['name'] = companies['name'].str.encode('utf-8').str.decode('ascii', 'ignore')
companies
```

```python
#decode the values of the field permalink

companies[['permalink']] = companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
companies



```

```python
#Decode the field company_permalink
rounds2[['company_permalink']] = rounds2[['company_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
rounds2
```

```python
#convert the values in the company permalink to lower case to make sure the values can be joined to create Master data frame
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
rounds2
```

```python
#convert the values in the permalink to lower case to make sure the values can be joined to create Master data frame

companies['permalink'] = companies['permalink'].str.lower()
companies
```

```python
#Create master_frame by merging the datas in the two sheets and filling the NaN values in raised_amount_usd  to 0 for ease of calculation
master_frame = pd.merge(rounds2, companies, left_on='company_permalink', right_on='permalink')
master_frame['raised_amount_usd'].fillna(0, inplace=True)
master_frame
```

```python
#data frame to create top3 english countries with investment type venture
countries = ['USA', 'GBR', 'IND']
top3_english = venture_type.loc[venture_type['country_code'].isin(countries)].sort_values(by = 'raised_amount_usd', ascending = False)
top3_english
```

```python
#get the primary sector from category list.

m = top3_english.category_list.str.split('|', n = 1, expand = True)
top3_english['main_sector'] = m[0]
top3_english
```

```python
#upload the mapping file
mapping = pd.read_csv('https://cdn.upgrad.com/UpGrad/temp/231dc91c-0642-470d-a362-29ddcd7142ce/mapping.csv')
mapping
```

```python
new = pd.melt(mapping, id_vars = ['category_list'], var_name ='primary_sector')
new
```

```python
#replace '0' with 'na' to clean the data
new['category_list']=new['category_list'].str.replace('0','na')
new
```

```python
#create sector analysis dataframe by merging the two data frames.
sector_analysis = pd.merge(top3_english,new, on = 'category_list', how = 'inner')
sector_analysis
```

```python
#apply the condition which has been provided in the case study
sector_analysis = sector_analysis [(sector_analysis['raised_amount_usd']>5000000)&(sector_analysis['raised_amount_usd']<15000000)]
sector_analysis
```

```python
#remove blanks from the primary sector
sector_analysis = sector_analysis[sector_analysis["primary_sector"] != "Blanks"]
sector_analysis
```

```python
#select where value=1 cause they will be the ones to be used in our analysis
sector_analysis = sector_analysis[sector_analysis["value"] == 1] 
sector_analysis
```

```python
#count of countries
sector_analysis.groupby('country_code')['value'].sum()
```
